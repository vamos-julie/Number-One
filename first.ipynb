```python
#histogram and normal probability plot
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Plot histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.histplot(df['Admission_grade'], kde=True, stat='density')

# Plot probability plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
stats.probplot(df['Admission_grade'], plot=plt)
```

```python
df_encoded = pd.get_dummies(df, columns=['Study_status'], prefix='', prefix_sep='')
df_encoded
```

```python
import matplotlib.pyplot as plt
corrmat = df_encoded.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = df_encoded.corr()

# Convert correlation matrix to absolute values
abs_corr_matrix = corr_matrix.abs()

# Select the top correlated columns
k = 10  # Number of variables for heatmap
cols = abs_corr_matrix.nlargest(k, 'Dropout')['Dropout'].index

# Calculate the correlation matrix for the selected columns
cm = np.corrcoef(df_encoded[cols].values.T)

# Create the heatmap
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
# Select the top correlated columns
k = 10  # Number of variables for heatmap
cols2 = abs_corr_matrix.nlargest(k, 'Graduate')['Graduate'].index

# Calculate the correlation matrix for the selected columns
cm2 = np.corrcoef(df_encoded[cols2].values.T)

# Create the heatmap
hm2 = sns.heatmap(cm2, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols2.values, xticklabels=cols2.values)
plt.show()
```

```python
# Select the top correlated columns
k = 10  # Number of variables for heatmap
cols3 = abs_corr_matrix.nlargest(k, 'Enrolled')['Enrolled'].index

# Calculate the correlation matrix for the selected columns
cm3 = np.corrcoef(df_encoded[cols3].values.T)

# Create the heatmap
hm3 = sns.heatmap(cm3, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols3.values, xticklabels=cols3.values)
plt.show()
```
